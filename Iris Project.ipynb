```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


plt.rcParams['font.size']=15
plt.rcParams['figure.figsize']=(15,7)
sns.set_style('darkgrid')
```

```python
data=pd.read_csv('Iris.csv')
data
```

```python
data.drop(['Id'],axis=1,inplace=True)
```

```python
sns.pairplot(data,hue='Species')
```

```python
sns.lmplot(x='PetalLengthCm',y='SepalLengthCm',hue='Species',data=data)
```

# Species-versicolor

```python
new_data=data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
new_df=data.Species=='Iris-versicolor'
```

```python
iris=new_data[new_df]
iris_df=iris.sort_values(by=['PetalLengthCm'],ascending=False)
iris_df
```

```python
sns.lmplot(x='SepalLengthCm',y='SepalWidthCm',data=iris_df)
```

```python
sns.lmplot(x='PetalLengthCm',y='PetalWidthCm',data=iris_df)
```
