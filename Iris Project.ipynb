# Linear-Regression

```python
sns.scatterplot(x=data.PetalLengthCm,y=data.SepalLengthCm,s=200,color='green')
```

```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
reg.fit(data[['PetalLengthCm']],data.SepalLengthCm)
```

```python
sns.scatterplot(data.PetalLengthCm,data.SepalLengthCm, color='g',s=130)
plt.plot(data.PetalLengthCm, reg.predict(data[['PetalLengthCm']]), color='red',linewidth=7)
plt.legend(['Regression_Line','DataPoints'])
```

```python
prediction = reg.predict(data[['PetalLengthCm']])
prediction
```

```python
data['Predicted_SepalLengthCm']=prediction
data
```

```python
new=data[['SepalLengthCm','Predicted_SepalLengthCm']]
new
```

```python
plt.plot(new['SepalLengthCm'],linewidth=3)
plt.plot(new['Predicted_SepalLengthCm'],linewidth=3)
plt.legend(new)
```

```python

sns.lmplot(x='SepalLengthCm',y='Predicted_SepalLengthCm',hue='Species',data=data)
plt.title('Actual_VS_Predicted Sepallengths')
```
