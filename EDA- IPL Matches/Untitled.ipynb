# To Perform EDA ON IPL Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['font.size']=15
plt.rcParams['figure.figsize']=(15,6)
sns.set_style('darkgrid')
```

```python
raw_data=pd.read_csv('matches.csv')
raw_data
```

```python
raw_data.drop(['id','date'],axis=1,inplace=True)
```

```python
raw_data
```

```python
raw_data.winner.value_counts()
```

```python
plt.pie(raw_data['winner'].value_counts(),labels=raw_data['winner'].value_counts().index,autopct='%1.1f%%')
```

```python
raw_data.venue.value_counts().head(10)
```

```python
plt.xticks(rotation=80)
plt.title('Hot FAvourite Venue\
')
sns.barplot(y=raw_data.venue.value_counts().head(10),x=raw_data.venue.value_counts().head(10).index)
```

```python
raw_data.player_of_match.value_counts().head(10)
```

```python
sns.barplot(x=raw_data.player_of_match.value_counts().head(10).index,y=raw_data.player_of_match.value_counts().head(10),palette='rocket')
```

```python
se=raw_data[['toss_winner','winner','player_of_match']]
se_df=raw_data.season==2019
```

```python
new=se[se_df]
new_df=new.sort_values(by=['toss_winner'],ascending=False)
new_df
```

```python
new_df.winner.value_counts()
```

```python
plt.xticks(rotation=60)
sns.barplot(x=new_df.winner.value_counts().index,y=new_df.winner.value_counts(),palette='magma')
```

```python
new_df.player_of_match.value_counts()
```
